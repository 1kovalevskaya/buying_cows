# Выбор коров для покупки с помощью моделей МО

**Описание проекта:** К нам обратился фермер, владелец молочного хозяйства «Вольный луг». Он хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма». Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже. Поэтому он просит нас разработать модель машинного обучения, которая поможет ему управлять рисками и принимать объективное решение о покупке. «ЭкоФерма» готова предоставить подробные данные о своих коровах. 
Вам нужно создать две прогнозные модели для отбора бурёнок в поголовье:
Первая будет прогнозировать возможный удой коровы (целевой признак Удой);
Вторая — рассчитывать вероятность получить вкусное молоко от коровы (целевой признак Вкус молока).
С помощью модели нужно отобрать коров по двум критериям:
средний удой за год — не менее 6000 килограммов;
молоко должно быть вкусным.

**Цель проекта:** Создать две прогнозные модели для отбора бурёнок в поголовье:
- Первая будет прогнозировать возможный удой коровы;
- Вторая — рассчитывать вероятность получить вкусное молоко от коровы.


**Ход исследования:**
1. Подготовка данных - загрузка и изучение общей информации из предоставленных датасетов.
2. Предобработка данных - поиск пропущенных значений, дубликатов, корректировка типов данных, заполнение пропусков, замена названий столбцов (приведение к нижнему регистру), поиск аномальных значений.
3. Исследовательский анализ данных - изучение распределений количественных и категориальных признаков в датасетах.
4. Обучение модели линейной регрессии для определения коров, приносящих молока не менее 6000 кг.
5. Обучение модели логистической регресии для определения коров, дающих вкусное молоко.
6. Выбор подходящих коров из предложенных фермеру. 

**Общий вывод:** С помощью моделей нужно отобрать коров по двум критериям:
- средний удой за год — не менее 6000 килограммов;
- молоко должно быть вкусным.

## Корреляционный анализ

```python
#вычисление матрицы корреляции
corr_matrix_0 = ferma_main.drop(['id'], axis=1).phik_matrix(interval_cols=num_col_ferma_main)

#визуализация матрицы корреляции 
plt.figure(figsize=(10, 5))
sns.heatmap(corr_matrix_0, annot=True, cmap='BuPu')
plt.title('Phi_K Correlation Matrix')
plt.show()
```

- Для признака `удой` мы можем говорить о сильной линейной взаимосвязи с параметрами `ЭКЕ`, `СПО` и `жирность`. Слабая линейная взаимосвязь присутствует с `протеином`. Нет линейной взаимосвязи с параметром `белок`;
- Для признака `ЭКЕ` наблюдается сильная линейная взаимосвязь с параметрами `удой`, `СПО` и `жирность`. Слабая взаимосвязь с параметром `протеин` и нет линейной взаимосвязи с параметром `белок`;
- Параметр `протеин` имеет слабую линейную взаимосвязь со всеми количественными признаками;
- Для признакак `СПО` наблюдается сильная линейная взаимосвязь с параметрами `удой`, `ЭКЕ` и `жирность`. Слабая корреляция с параметром `протеин` и никакой взаимосвязи с параметром `белок`;
- Параметр `жирность` имеет сильную взаимосвязь с параметрами `удой`, `ЭКЕ` и `СПО`, слабую - с `протеин` и никакой взаимосвязи с параметром `белок`.
- `Белок` имеет лишь слабую связь с параметром `протеин` и никакой корреляции с другими признаками.

```python
#создание функции для рассчета коэффициента корреляции и построения scatter plot
def correlation(category, column):
    i = 1
    plt.figure(figsize=(15,3))
    for name in ferma_main[category].unique():
        l = len(ferma_main[category].unique())
        coff_corr = ferma_main[ferma_main[category] == name][column].corr(ferma_main[ferma_main[category] == name]['удой'])
        plt.subplot(1, l, i)
        plt.title(f'Удой от {column} для {name}\n Коэф-т корреляции Спирмена: {round(coff_corr, 3)}')
        plt.xlabel(column)
        plt.ylabel('удой')
        plt.scatter(x=ferma_main[ferma_main[category] == name][column], y=ferma_main[ferma_main[category] == name]['удой']);
        i += 1
```

```python
num_col_ferma_main = ['ЭКЕ', 'протеин', 'СПО', 'жирность', 'белок'] 
cat_col_ferma_main = ['порода', 'пастбище', 'порода_папы', 'вкус', 'возраст']
for num in num_col_ferma_main:
    for cat in cat_col_ferma_main:
        correlation(cat, num)
```

- Параметр `удой` имеют сильную взаимосвязь с параметром `ЭКЕ` вне зависимости от породы коровы, пастбища, на котором она паслась и вкуса молока.
- Однако наблюдается небольшая разность в `удое` в зависимости от `породы отца` - для породы Соверин удой в среднем меньше, чем от породы Айдиал. 
- Существенная разница в зависимости `удоя` от `ЭКЕ` наблюдается для коров разного возраста - более старшие коровы дают больше молока и имеют более сильную взаимосвязь между `удоем` и `ЭКЕ`. А молодые коровы вообще не имеют взаимосвязи с целевым признаком.

Также наблюдается нелинейная взаимосвязь между данными параметрами.

- Зависимость `удоя` от `протеина` не отличается для коров разных пород, не зависит от типа пастбища и вкуса молока. 
- Однако наблюдается небольшая разность в `удое` в зависимости от `породы отца` - для породы Соверин удой в среднем меньше, чем от породы Айдиал. 
- Существенная разница в зависимости `удоя` от `протеина` наблюдается для коров разного возраста - более старшие коровы дают больше молока и имеют более сильную взаимосвязь между `удоем` и `ЭКЕ`. Молодые коровы вообще не имеют взаимосвязи.
Наблюдается слабая взаимосвязь между `удоем` и `протеином`. 

- В целом, на все зависимости `удоя` от количественных признаков не влияют `порода коровы`, `тип пастбища` и `вкус молока`. 
- Есть небольшая разница в зависимости от `породы отца` (интересно почему? было бы логичнее, если бы от породы самой коровы).
- Большая разница наблюдается для коров старше 2 лет и младше 2 лет - то есть `возраст коровы` сильно влияет на зависимость удоя от количественных признаков.

В целом коэффициенты корреляции  между `удоем` и количественными признаками для отдельных групп категориальных признаков похожи на коэффициенты корреляции без разделения на категории.

Параметр СПО лучше всего привести к категориальному - больше 0.91 и меньше. 

## Обучение модели линейной регрессии

```python
#создание функции для расчета метрик для регресионной модели

def metrics(y_true, y_pred):
    MSE = mean_squared_error(y_true, y_pred)
    RMSE = mean_squared_error(y_true, y_pred, squared=False)
    MAE = mean_absolute_error(y_true, y_pred)
    R2 = r2_score(y_true, y_pred)
    print(f"R2 = {R2:.3f}")
    print(f"MSE = {MSE:.0f}")
    print(f"RMSE = {RMSE:.0f}")
    print(f"MAE = {MAE:.0f}")    
```

```python
#создание фукнции для построения гистограммы распределения и дисперсии остатков

def residuals(y_true, y_pred):
    residuals = y_true - y_pred
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    axes[0].hist(residuals)
    axes[0].set_title('Гистограмма распределения остатков')
    axes[0].set_xlabel('Остатки')
    axes[1].scatter(x=y_pred, y=residuals)
    axes[1].set_xlabel('Предсказания модели')
    axes[1].set_ylabel('Остатки')
    axes[1].set_title('Анализ дисперсии')
    plt.show()
```

Остальные количественные признаки в датасете ferma_main либо имеют слабую линейную взаимосвязь с целевым признаком, либо имеют обратную причинно-следственную связь - наврядли характеристики самого молока влияют на удой, скорее всего наоборот.

```python
#создание функцию для вывода метрик

def metric_log(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    print(f"accuracy = {accuracy:.3f}")
    print(f"precision = {precision:.3f}")
    print(f"recall = {recall:.3f}")  
```

```python
#функция для вывода матрицы ошибок
def error_matrix(y_test, y_pred):
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
    plt.ylabel('Истинное значение')
    plt.xlabel('Предсказанное значение');
```

### Создание модели логистической регрессии

```python
data.head()
```

```python
#Разделение входных и целового признаков 

num_col_data = ['ЭКЕ', 'СПО', 'ЭКЕ_new', 'протеин', 'жирность', 'белок']
cat_col_data = ['порода', 'пастбище', 'возраст', 'порода_папы', 'СПО_new', 'имя_папы']

X = data.drop('вкус', axis=1)
y = data['вкус']
```

```python
#создание тренировочной и тестовой выборок

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=RANDOM_STATE) 
```

```python
#масштабирование количественных признаков

scaler_4 = StandardScaler() #выбор метода масштабирования

X_train_scaled_4 = scaler_4.fit_transform(X_train[num_col_data])
X_test_scaled_4 = scaler_4.transform(X_test[num_col_data]) 
```

```python
#One-Hot encoding категориальных признаков

encoder_4 = OneHotEncoder(drop='first', sparse=False) #выбор метода кодирования

X_train_ohe_4 = encoder_4.fit_transform(X_train[cat_col_data])
X_test_ohe_4 = encoder_4.transform(X_test[cat_col_data])

encoder_col_names_4 = encoder_4.get_feature_names() #список новых столбцов

X_train_ohe_4 = pd.DataFrame(X_train_ohe_4, columns=encoder_col_names_4)
X_test_ohe_4 = pd.DataFrame(X_test_ohe_4, columns=encoder_col_names_4)

X_train_scaled_4 = pd.DataFrame(X_train_scaled_4, columns=num_col_data)
X_test_scaled_4 = pd.DataFrame(X_test_scaled_4, columns=num_col_data)

X_train_4 = pd.concat([X_train_ohe_4, X_train_scaled_4], axis=1)
X_test_4 = pd.concat([X_test_ohe_4, X_test_scaled_4], axis=1)
```

Поскольку для фермера важно, чтобы молоко было вкусным, то нам нужно избегать ошибки 1-го рода, то есть ложноположительных результатов. Для минимизации таких ошибок необходимо улучшать метрику precision. 

```python
# нахождение оптимального порога классификации для увеличения метрики precision
y_proba = clf.predict_proba(X_test_4)[:,1]

df = pd.DataFrame(zip(y_test, y_proba), columns = ['y_valid', 'y_proba']).sort_values(by='y_proba',ascending=False)


thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 10,endpoint=False)]

columns = []

for i in thresholds:
    columns.append('y_pred_'+str(i))
    df['y_pred_'+str(i)] = df['y_proba'].apply(lambda x: 1 if x > i else 0)
    print(i, precision_score(df['y_valid'], df['y_pred_'+str(i)]))
```

Оптимальным порогом можно выбрать значение равное 0.82. Рассчитаем все метрики заново с новым порогом классификации.

```python
i = 0.82
columns.append('y_pred_'+str(i))
df['y_pred_'+str(i)] = df['y_proba'].apply(lambda x: 1 if x > i else 0)
print('precision:', precision_score(df['y_valid'], df['y_pred_'+str(i)]))
print('recall:', recall_score(df['y_valid'], df['y_pred_'+str(i)]))
```

**Была обучена модель линейной регрессии 3-мя способами:**
- Входные признаки были взяты в неизменном виде. Анализ дисперсии остатков данной модели показазал, что модель плохо работает при больших значениях целевого признака. Также остатки распределены не совсем равномерно - отрицательные остатки встречаются чаще, что говорит о том, что модель часто завышает значения целевого признака. R2=0.748, MSE=51306 кг^2, RMSE=227 кг, MAE=181 кг. Модель предсказывает около 75% значений.
- Была скорректирована нелинейная зависимость количества получаемого молока и таких параметров как ЭКЕ и СПО. Параметр ЭКЕ был возведен в квадрат, а параметр СПО переведен в категориальный. Распределение остатков стало похоже на нормальное, дисперсия в области больших значений целевого признака всё также оставляет желать лучшего - модель плохо предсказывает большие значения, Метрики MSE, RMSE, MAE уменьшились (MSE=42009 кг^2, RMSE=2105 кг, MAE=160 кг), что говорит о том, что дополнительное преобразование входных признаков позволило несколько улучшить качество модели. Модель предсказывает около 79% значений - этот показатель немного повысился.
- Добавлен новый категориальный параметр - имя папы. Распределение остатков похоже на нормальное, дисперсия практически одинаковая на всём диапазоне значений целевого признака, метрики MSE, RMSE, MAE уменьшились (MSE=35316 кг^2, RMSE=188 кг, MAE=145 кг), что говорит о том, что дополнительное преобразование входных признаков позволило несколько улучшить качество модели. Модель предсказывает около 83% значений - этот показатель повысился по сравнению с другими моделями.

Из трех представленных моделей согласно значениям метрик была выбрана 3-я модель. Согласно этой модели, все представленные на продажу коровы будут приность более 6000 кг молока в среднем в год. 

**Была обучена модель логистической регрессии:**
- Во время обучения данной модели необходимо было минимизировать ошибки 1-го рода, то есть ложноположительные результаты. Таким образом необходимо было повысить метрику precision.
- На исходных данных precision равнялась 0.626. С помощью изменения порога классификации (стал равен 0.82) удалось повысить метрику до 1. 
- С новым порогом классификации ожидается, что 6 коров будут давать вкусное молоко.
